```python
briefings_df = pd.read_csv('../data/whtfb/all_briefings.csv')

# REMOVE FROM THIS SCRIPT, ANALYSIS IS WHERE I HAD THIS BEFORE - ADD TO scrape???
briefings_df = briefings_df[briefings_df["speaker"].str.contains('Speaker ')==False] # check error on first run for this

briefings_df['speaker'].replace(regex={r'.*Trump.*': 'Donald Trump', 
                                       r'.*Pence.*': 'Mike Pence',
                                       r'.*Fauci.*': 'Dr. Anthony Fauci',
                                       r'.*Birx.*': 'Dr. Deborah Birx',
                                       r'.*Berks.*': 'Dr. Deborah Birx',
                                       r'.*Pompeo.*': 'Mike Pompeo',
                                       # r'.*Peter.*': 'Peter Navarro', check this one
                                       r'.*Novarro.*': 'Peter Navarro',
                                       r'.*Report.*': 'Reporter',
                                       r'.*Crowd.*': 'Reporter',
                                       r'.*Group.*': 'Reporter',
                                       r'.*Audience Member.*': 'Reporter',
                                       r'.*Surgeon General.*': 'Surgeon General Jerome Adams',
                                       r'.*Jerome Adams.*': 'Surgeon General Jerome Adams',
                                       r'.*Mnuchin.*': 'Secretary Steve Mnuchin',
                                       r'.*Speaker .*': 'Unnamed Speaker'}, 
                                inplace = True)

```

```python
data = pd.DataFrame([])

for text in briefings_df['text']:
    paragraph_words = text.split()
    paragraph_emos = emolex_words[pd.DataFrame(emolex_words.word.tolist()).isin(paragraph_words).any(1)].mean()
    data = data.append(paragraph_emos, ignore_index=True)

briefings_df = briefings_df.join(data)
briefings_df

# word column didn't apear when i ran iloc to subset the first x rows??
# emo scores are getting scaled down for counts of the negative/positive words??
```

```python
# adjust date range??? see subset below
trump_feels = briefings_df[briefings_df['speaker']=='Donald Trump']
trump_feels_by_date = trump_feels[130:].groupby(['date']).mean().reset_index()
trump_feels_by_date = trump_feels_by_date.drop(columns=['word'])
trump_feels_by_date = trump_feels_by_date.melt('date', var_name='emo', value_name='value')
trump_feels_by_date

alt.Chart(trump_feels_by_date).mark_line().encode(
    x='date:T',
    y='value:Q',
    color='emo:N'
).properties(width=800, height = 600)
```
